# 오후 Q&A

### 순전파, 역전파
- **순전파**
    - 입력층 -> 출력층
    - 가중치, bias 학습 과정
- **역전파**
    - 실제값과 예측값의 오류를 최소화하기 위하여 가중치를 업데이트 하는 방식으로 편미분을 사용하는 신경망 학습 알고리즘
    - 실제값과 예측값 사이의 오차를 최소화
    - 가중치와 편향을 학습한다.
    - 입력값, 실제값 고정
    - 예측값 -> 가중치, 편향 학습하는 수 밖에 없습니다.
    

### 가중치
- 입력값의 영향력 비중
- 입력이 결과 출력에 주는 영향도를 조절하는 변수
- 필터 역할

### 편향(Bias)
- 노드가 얼마나 쉽게 활성화될 지 조절하는 역할

### 뉴런
- 노드
- 정보를 받아서 처리하는 부분
- 외부로부터 신호를 받아들임
- 입력값 연산 수행 내보내는

### Input layer 입력층
- 신호(데이터)를 받아들이기만 한다.
- 연산은 수행하지 않음.
- input layer의 노드 수 = feature 수

### output layer 출력층
- 최종 결과 출력
- output layer의 노드 수 = 타겟 클래스 수
- 이진분류: 노드 수 = 1, 활성함수 = sigmoid
- 다중분류(5개 클래스) : 노드 수 = 5, 활성함수 = softmax

### hidden layer 은닉층
- 연산이 수행되는 곳. 사람이 연산 수행 과정을 들여다 볼 수 없음.
- 히든 레이어의 노드 수 = 정해지지 않음. 그때그때 달라요. 내가 값을 찾아야 함.

### 퍼셉트론
- 신경망 구조의 최소단위
- input 여러 개, output 1개


### learning rate 학습률
- 학습 결과가 얼마나 반영할 지 조절
- 크면 : 발산
- 작으면 : 최적점에 이르지 못하고 학습이 종료될 수 있음

### batch
- 한번 학습할 때 사용할 데이터의 개수
- batch 사이즈가 크면 : 메모리 많이 사용. 속도 저하  
    왜? 병렬처리 방식
- 작으면 : 학습 시간이 오래 걸림

### epoch
- 전체 데이터가 학습되는 횟수
- 크면 : 과적합, 학습 오래 걸림  
    과적합 방지 : early stopping
- 작으면 : 과소적합

1,000개의 데이터, batch size 100
- 총 학습은? 10번
- epoch 2일 때 총 학습은? 20번

### 손실 함수
- 실제값과 예측값의 차이
- MAE, MSE, cross entropy 등
- 모델 학습할 때 사용

### 성능 측정
- metric

### overfitting 방지하는 법
- weight decay
    - L1, L2
    - 손실함수에 가중치의 합을 포함시킨다
    - lamda
    - L1 : 절대값
    - L2 : 제곱
    - L1 vs L2 차이 
        - L1은ㄴ 중요하지 않은 특성은 0으로 수렴
        - L1은 가중치가 가장 큰 것만 남고, L2는 조금이라도 반영 됨
        
        
- dropout
    - 노드의 성능 향상을 위해
    - 랜덤으로 노드를 끈다.
    - 학습할 때만 끄고, 출력할 때는 포함된다.
    
    
- early stopping


- constraints
    - 가중치가 가질 수 있는 값의 범위를 정해주는 것


### 활성화함수, 활성함수 activation function
- 가중치, 입력된 값, 편향 -> 보정 -> 출력
- sigmoid, softmax
- relu, tanh, leaky relu, step, elu

### Cross validation
- 과적합 방지
- 하이퍼파라미터 튜닝
- 데이터의 수가 적더라도 train/test 분리 가능

### train/test 분리 이유
- data leakage 발생 가능
- 평가가 제대로 이뤄지지 않는다.
